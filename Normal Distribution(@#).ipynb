## Normal Distribution

```python
#Normal distribution is defined by mean and variance
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 0
sigma = 1
#Return evenly spaced values within a given interval (start,stop,step).
x = np.arange(-5,5,0.1)
#Generate normal distribution that complies with mu an sigma.
y = norm.pdf(x,mu,sigma)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('density')
plt.show()
```
